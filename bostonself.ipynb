## Data Exploration

```python
boston=pd.read_csv('boston_house_prices.csv')
boston.head(200)
```

```python
boston.columns=boston.iloc[0]
```

```python
boston=boston.iloc[1:]
boston=boston.reset_index(drop=True)
boston
```

```python
boston.info()
```

```python
boston=boston.apply(pd.to_numeric)
boston.info()
```

```python
boston.isnull().sum()
```

```python
boston.describe().T
```

```python
sns.pairplot(x_vars=boston.columns,y_vars='MEDV',data=boston,kind='scatter',plot_kws={'alpha':0.4})
plt.show()
```

```python
sns.lmplot(x='RM',y='MEDV',data=boston,scatter_kws={'alpha':0.3})
sns.lmplot(x='LSTAT',y='MEDV',data=boston,scatter_kws={'alpha':0.3})
sns.lmplot(x='AGE',y='MEDV',data=boston,scatter_kws={'alpha':0.3})
sns.lmplot(x='NOX',y='MEDV',data=boston,scatter_kws={'alpha':0.3})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=boston.drop('MEDV', axis=1)
y=boston['MEDV']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
sns.scatterplot(x=y_test,y=y_pred)
plt.ylabel('Predictions')
```
